```python
plt.figure(figsize=(5,7))                               # more than 100 matches were there, where the teams won b/w 0-20 runs
plt.hist(batting_first["win_by_runs"])
plt.title("Win By Runs")
plt.xlabel("Runs")
plt.show()
```

```python
# Finding out the no. of wins with respect to each team after batting first
batting_first["winner"].value_counts()
```

```python
# Making a bar-plot for top 3 teams with most wins after batting first
plt.figure(figsize=(6,6))
plt.bar(list(batting_first["winner"].value_counts()[0:3].keys()),list(batting_first["winner"].value_counts()[0:3].values),color=["Blue","Yellow","Red"])
plt.title("Top 3 teams")
plt.xlabel("Teams")
plt.ylabel("No. of Wins")
plt.show()
```

In pie chart first parameter would be numerical and second would be categorical.

```python
plt.pie(list(batting_first["winner"].value_counts().values),labels=list(batting_first["winner"].value_counts().keys()),autopct="%0.1f%%"
plt.show()
```
